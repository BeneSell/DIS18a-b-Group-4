```python
from Bio import Entrez
import re

Entrez.email = "emirhan.acikgoez@hotmail.com"  # Always tell NCBI who you are

# Search PubMed for articles with "GC content bacterial" in the title
term = "GC content bacterial[Title]"
handle = Entrez.esearch(db="pubmed", term=term)
record = Entrez.read(handle)
ids = record["IdList"]

# Retrieve the abstracts for the articles
handle = Entrez.efetch(db="pubmed", id=ids, rettype="abstract", retmode="xml")

records = Entrez.read(handle)

# Select articles with "GC", "content", and "bacterial" in the title and 1% in the abstract
abstracts_with_one_perc = []
for record in records["PubmedArticle"]:
    try:
        temp_record = record["MedlineCitation"]["Article"]
        title = temp_record["ArticleTitle"]
        abstract = temp_record["Abstract"]["AbstractText"][0]
        if "GC" in title and "content" in title and "bacterial" in title and "%" in abstract:
            matches = re.findall(r"\%", abstract)
            if len(matches) == 1:
                doi = temp_record.get('ELocationID', None)
                if isinstance(doi, list):
                    doi = doi[0]
                abstracts_with_one_perc.append((temp_record['MedlineCitation'], temp_record['Article'], doi, title))
    except:
        pass

# Print the selected records
for record in abstracts_with_one_perc:
    medline_citation = record[0]
    article = record[1]
    doi = record[2]
    title = record[3]
    
    print("MedlineCitation:", medline_citation)
    print("Abstract:", article['Abstract']['AbstractText'][0])
    print("DOI:", doi)
    print("Title:", title)
    print("--------------------------------------------------------")
```
